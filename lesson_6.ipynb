```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x**3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=False)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax1.set_title('x**2')
ax1.set_xlim([-3, 3])
ax1.set_ylim([-0.1, 5])
ax2.plot(x, x3)
ax2.set_title('x**3')
ax2.set_ylim([-10, 10])
ax3.plot(x, y)
ax3.set_title('y')
ax3.grid()
ax4.plot(x, z)
ax4.set_title('z')
fig.set_size_inches(7, 6)
plt.subplots_adjust(wspace=0.5, hspace=0.5)
```
